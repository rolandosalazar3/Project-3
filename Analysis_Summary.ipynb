```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
%matplotlib inline
```

```python
data_path = "university_data.csv"
university_df = pd.read_csv(data_path).set_index("Name")
```

```python
#SUMMARY OF STATISTICS

# Calculate the total number of unique schools
print(len(university_df))

# Calculate the average number of undergraduate students
print(university_df["Undergrad_Enrollment"].mean())

# Calculate the average tuition
print(university_df["TuitionFees"].mean())
```

```python
#Use .loc to narrow down the data to show only the schools with the most undergraduate students
stu_df = university_df.groupby(["Undergrad_Enrollment"])
stu_df.max(10)
```

```python
# Find the states with the most universities
states_avg = university_df["state_code"].value_counts()
states_avg.head(15)
```

```python
# Find the states with the most universities
city_avg = university_df["city"].value_counts()
city_avg.head(15)
```

```python
#Use .loc to narrow down the data to show only the top 10 schools
top_schools_df = university_df.loc[university_df["Rank_USNewsWorldReport"]<= 10,:]
top_schools_df
```
